Write a program to swap two numbers in Python with the help of a temporary variable

```
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))
temp  = num1
num1 = num2
num2 = temp
print("Swap numbers are", num1,num2)
```

Write a program to swap two numbers in Python without using a temporary variable

```
num1 = int(input("Enter first number:"))
num2 = int(input("Enter first number:"))
num1,num2 = num2,num1
print("Swap numbers are", num1,num2)
```

Write a program to convert fahrenheit to kelvin and celsius both

```
F_temp = int(input("Enter temperature in Fahrenheit(F):"))
kelvin_temp = ((F_temp - 32) / 1.8) + 273.15
celsius_temp = (F_temp - 32) / 1.8
print("Temperature from Fahrenheit to Kelvin and Celsius are ",kelvin_temp,celsius_temp)
```
